```python
from curves_to_parameters import Data
from curves_to_parameters import DeepL
from curves_to_parameters import Viz
```

```python
sin_gen = Data.Sinusoids_Generator()
```

```python
batch_size = 32
epochs = 15
model = DeepL.get_compiled_model(optimizer = 'adam', loss = "mean_squared_error")
X, Y = sin_gen.generate_training_dataset()
```

```python
history, losses = DeepL.train_keras(batch_size, epochs, model, X, Y)
```

```python
mr = DeepL.Model_Results('LSTM_64_2', losses = losses)
```

```python
mr.save_mr(model)
```
